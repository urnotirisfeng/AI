### Problem 1
Write a brief description of each of these agent types and provide an example of where we’ve seen each of these types in an assignment or lecture example:
- Reflex
- Model-based
- Goal-based
- Utility-based
- Learning


* **Reflex**: This agent selects an action based on the present, does not consider future consequences and encode preferences as a function of the percepts and action. For example, in the robot vacuum example, if the floor is dirty, the robot will clean it. If not, then the robot will move in the opposite direction. 


* **Model-based**: This agent selects an action based on the past and present, does not consider future consequences and maintains an internal state to keep track of what the world looks like. For example, the vacuum agent knows what it cleaned, but still does not search for the most optimal path. 


* **Goal-based**: This agent selects an action based on the past, present, and future. There is also a goal that the agent work towards. Decisions made by the agent are also based on hypothesized consequences of proposed actions. For example, the vacuum agent may be designed to clean the room in the fewest number of actions. 


* **Utility-based**: This agent selects an action based on the past, present, and future. Utility quantifies the benefit of various actions. For example, the agent can assign a socre to a solution and chooses the solution with the maximum score. 


* **Learning**: This agent separates percepts into a component of reward or penalty. For example, finding out if an email is spam by analysing the words in the email.

### Problem 4
A desk lamp produced by The Luminar Company was found to be defective (D). There are three factories (A, B, C) where such desk lamps are manufactured. A Quality Control Manager (QCM) is responsible for investigating the source of found defects. This is what the QCM knows about the company's desk lamp production and the possible source of defects: 

| Factory| % of total production | Prob. of defective lamps |
| ------ | --------------------- | ------------------------ |
| A	| 0.35| 0.015 |
| --|-----| ----- |
| B	| 0.35| 0.010 |
| --|-----| ----- |
| C	| 0.30| 0.020 |
| --|---- | ----- |

#### Question
A defective lamp is returned to the company. What is the probability the lamp was produced in factory A?

Let A be the event that the lamp was produced in factory A. 

Let D be the event that the lamp was defective. 

P(D) = P(D | A) * P(A) + P(D | B) * P(B) + P(D | C) * P(C)
= 0.015 * 0.35 + 0.010 * 0.35 + 0.020 * 0.30
= 0.01475

P(A | D) = P(A, D)/P(D) = P(D | A) * P(A) / P(D) = 0.015 * 0.35 / 0.01475 = 0.356
